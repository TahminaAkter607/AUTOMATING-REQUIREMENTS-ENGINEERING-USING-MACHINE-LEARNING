```python
projects=[];
with open(filename2,'r') as csvfile:
  datareader=csv.reader(csvfile)
  for row in datareader:
    if(row[1]==""):
      break
    projects.append(row[1])
print(projects)
```

```python
matched_project_index=[]
x="Online delivery system"
user_input=input("Enter system name  :  ")
print(user_input)
projects[0]=user_input
sentence_embeddings = model.encode(projects)
sentence_embeddings.shape
from sklearn.metrics.pairwise import cosine_similarity
project_name_similarityScore=cosine_similarity(
    [sentence_embeddings[0]],
    sentence_embeddings[1:]
).tolist()
#print(project_name_similarityScore[0])
for i in range(len(project_name_similarityScore[0])):
  if project_name_similarityScore[0][i]>=0.80:
    matched_project_index.append(i+1)
if(len(matched_project_index))<2:
  for i in range(len(project_name_similarityScore[0])):
    if project_name_similarityScore[0][i]>=0.70:
     matched_project_index.append(i+1)
if(len(matched_project_index))<2:
  for i in range(len(project_name_similarityScore[0])):
    if project_name_similarityScore[0][i]>=0.70:
     matched_project_index.append(i+1)
if(len(matched_project_index))<2:
  for i in range(len(project_name_similarityScore[0])):
    if project_name_similarityScore[0][i]>=0.60:
     matched_project_index.append(i+1)
if(len(matched_project_index))<2:
  for i in range(len(project_name_similarityScore[0])):
    if project_name_similarityScore[0][i]>=0.50:
     matched_project_index.append(i+1)
if(len(matched_project_index))<2:
  print("No requirements Found")
#print(matched_project_index)
```
